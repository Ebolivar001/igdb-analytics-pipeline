```python
from pyspark.sql import SparkSession
from pyspark import SparkContext

SpSession = SparkSession \
        .builder \
        .appName("Demo Spark") \
        .getOrCreate()
```

```python
SpContext = SpSession.sparkContext
```

```python
data = SpSession.read.csv('cars.csv', header=True, sep= ";")
data.show(5)
```

```python
data.printSchema()
```

```python
data.columns
```

```python
#Metodo 1

data.select(data.Car).show(truncate=False) ##To show full data and not piece of it
```

```python
#Metodo 2

data.select(data['car']).show(truncate=False)
```

```python
from pyspark.sql.functions import col
```

```python
#Metodo 3

data.select(col('car')).show(truncate=False)
```

```python
##Seleccion multiple

data.select(data.Car, data.Cylinders).show(truncate=False)
```

```python
from pyspark.sql.functions import lit
```

```python
##agregar nueva columna

df = data.withColumn('FirstColumn',lit(1))
df.show(5,truncate=False)
```

```python
##agregar nuevas columnas

df = data.withColumn('SecondColumn',lit(2))\
         .withColumn('ThirdColumn',lit(3))
df.show(5,truncate=False)
```

```python
##Group by

df.groupBy('Horsepower','Origin').count().show(5)
```

```python
#eliminar columnas

df = df.drop('SecondColumn')
```

```python
df
```

```python
df.orderBy('Model',ascending=False).show(truncate=False)
```

```python
df.groupBy('Car').count().orderBy('count',ascending=False).show(5)
```

```python
#Filtrado de datos

total_count=df.count()
print("Total Records:",total_count)
```

```python
europa_count = df.filter(col('Origin')=='Europe').count()
print("Total Records:",europa_count)
```

```python
df.filter(col('Origin')=='US').show(truncate=False)
```

```python
europa_count = df.filter(col('Origin')=='US').count()
print("Total Records:",europa_count)
```

```python
total_count=df.count()
print("Total Records:",total_count)

us_count = df.filter((col('Origin')=='US')&(col('Horsepower')=='170.0')).count()
print("Total Records:",us_count)

df.filter((col('Origin')=='US')&(col('Horsepower')=='170.0')).show(truncate=False)
```
